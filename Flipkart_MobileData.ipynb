```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

Product_name = []
Price = []
Description = []
Reviews = []

for i in range(2,12):
    url = "https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r = requests.get(url)
    # print(r)

    soup = BeautifulSoup(r.text, "lxml")
    box = soup.find("div", class_= "_1YokD2 _3Mn1Gg")

    names = box.findAll("div", class_="_4rR01T")

    for i in names:
        name = i.text
        Product_name.append(name)


    prices = box.findAll("div", class_="_30jeq3 _1_WHN1")
    for i in prices:
        name = i.text
        Price.append(name)

    desc = box.findAll("ul", class_="_1xgFaf")
    for i in desc:
        name = i.text
        Description.append(name)


    review = box.findAll("div", class_="_3LWZlK")
    for i in review:
        name = i.text
        Reviews.append(name)

        
# Filling missing values with placeholders
max_length = max(len(Product_name), len(Price), len(Description), len(Reviews))
Product_name += [''] * (max_length - len(Product_name))
Price += [''] * (max_length - len(Price))
Description += [''] * (max_length - len(Description))
Reviews += [''] * (max_length - len(Reviews))

# Create DataFrame
df = pd.DataFrame({"Product Name": Product_name, "Prices": Price, "Description": Description, "Review": Reviews})
print(df)

# Save DataFrame to Excel
df.to_csv(r"/Users/pranav/Downloads/Flipkart_Mobile_Data.csv")
```
